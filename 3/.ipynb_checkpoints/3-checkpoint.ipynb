```python
def removingOutliers(columnName, df):
    Q1 = df[columnName].quantile(0.25)
    Q3 = df[columnName].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    df = df[(df[columnName] >= lower_limit) & (df[columnName] <= upper_limit)]
```

```python
plt.figure(figsize = (6, 8))
sns.boxplot(y = df['PetalWidthCm'])
plt.ylabel("Petal Width")
plt.title('Boxplot of Petal Width')
plt.show()

removingOutliers(columnName = 'PetalWidthCm', df = df)

plt.figure(figsize = (6, 8))
sns.boxplot(y = df['PetalWidthCm'])
plt.ylabel("Petal Width")
plt.title('Boxplot of Petal Width')
plt.show()
```
